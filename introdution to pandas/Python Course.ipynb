```python
#CALCULATOR
def sum(a,b):
    x=a+b
    return x
def subtract(a,b):
    x=a-b
    return x
def div(a,b):
    x=a/b
    return x

def mul (a,b):
    x=a*b
    return x
def exp (a,b):
    x=a*b
    return x


print(sum(5,3))
print(subtract(5,3))
print(div(10,5))
print(mul(5,3))
print(exp(2,2))



   
```
