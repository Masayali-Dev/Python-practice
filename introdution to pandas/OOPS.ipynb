```python
class Student:
    def __init__(self, name, course, attendance, phone_number):
        self.name = name
        self.course = course
        self.attendance = attendance
        self.phone_number = phone_number

    def display_details(self):
        print(f"name          : {self.name}")
        print(f"course        : {self.course}")
        print(f"attendance    : {self.attendance}")
        print(f"phone_number  : {self.phone_number}")

    def is_present(self):
        return self.attendance.lower() == "present"


student = Student(
    "Sumit",
    "Data analytics",
    "Absent",
    "24465767889"
)

student.display_details()

if student.is_present():
    print("Student attended today's class.")
else:
    print("Student was absent today.")
```

Encapsulation in Python: Restricting direct access to attributes & methods 
 
Encapsulation restricts access to certain attributes or methods to protect the data and 
enforce controlled access. 

```python
class Student: 
    def __init__(self, name, grade, percentage): 
        self.name = name 
        self.grade = grade 
        self.__percentage = percentage  # Private attribute (hidden)

    def get_percentage(self):  # Public method to access the private attribute 

        return self.__percentage
    
  
```

```python
 # Creating a student object 
Elite = Student("Sayali", 10, 98) 
 
# Accessing the private attribute using the public method 
print(f"{Elite.name}'s percentage is {Elite.get_percentage()}%.")   

```

Inheritance in Python: Reusing Parent’s prop & methods 
 
Inheritance (parent-child), allows one class (child) to reuse the properties and methods of 
another class (parent). This avoids duplication and helps in code reuse. 

```python
class Student: 
    def __init__(self, name, grade, percentage): 
        self.name = name 
        self.grade = grade 
        self.percentage = percentage 
    def student_details(self):  # method  
        print(f'{self.name} is in {self.grade} grade with {self.percentage}%')


   
```

```python
class GraduateStudent(Student):  
               # GraduateStudent inherits from Student 
    def __init__(self, name, grade, percentage, stream): 
        super().__init__(name, grade, percentage)       # Call parent class initializer 
        self.stream = stream                    # New attribute specific to GraduateStudent 
    
    def student_details(self): 
        super().student_details() 
        print(f"Stream: {self.stream}")
```

```python
Elite = Student("Sayali",10,93)
Elite.student_details()
```

```python
# Create a graduate student 
Elite = GraduateStudent("Sayali", 12, 94, "PCM")  # Sayali is in 12 grade with 94% 
 
Elite.student_details()  # Stream: PCM        
```

Polymorphism in Python: Same method but different output 
 
Polymorphism allows methods in different classes to have the same name but behave 
differently depending on the object.

```python
class GraduateStudent(Student): 
    def student_details(self):  # Same method as in parent class   
     print(f"{self.name} is a graduate student from final year.") 
   
```

```python
# Polymorphism in action
Group_parent = Student("Sayali", 10, 98)
Group_parent.student_details()  # Output: Madhav is in 10 grade with 98
```

```python
Group_child= GraduateStudent("Sumit", 12, 99, "PCM")
Group_child.student_details()
```
