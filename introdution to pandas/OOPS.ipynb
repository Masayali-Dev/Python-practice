```python
class Student:
    def __init__(self, name, course, attendance, phone_number):
        self.name = name
        self.course = course
        self.attendance = attendance
        self.phone_number = phone_number

    def display_details(self):
        print(f"name          : {self.name}")
        print(f"course        : {self.course}")
        print(f"attendance    : {self.attendance}")
        print(f"phone_number  : {self.phone_number}")

    def is_present(self):
        return self.attendance.lower() == "present"


student = Student(
    "Sumit",
    "Data analytics",
    "Absent",
    "24465767889"
)

student.display_details()

if student.is_present():
    print("Student attended today's class.")
else:
    print("Student was absent today.")
```

```python
class BankAccount:
    def __init__(self, customer_name,balance):
       self.customer_name = customer_name
       self.balance = balance

    def deposit(self,amount):
        self.balance += amount
        print(f"₹{amount} deposited.") 

    def withdraw(self,amount):
        if amount<= self.balance:
            self.balance -= amount
            print(f"₹{amount} withdrawn.")
        else:
            print("Insufficient Balance")

    def check_balance(self):
            print(f"Current Balance: ₹{self.balance}")          



account = BankAccount("Sayali",10000)

account.deposit(5000)
account.withdraw(2000)
account.check_balance()




```

```python
class Machine:
    def __init__(self, machine_id):
        self.machine_id = machine_id
        self.production_count = 0

    def produce(self,quantity):
        self.production_count += quantity
        print(f"produced{quantity}units")

    def machine_status(self):
        print(f"machine{self.machine_id}is running")

    def total_production(self):
        print(f"total production: {self.production_count}")


machine =Machine("M101")        
              
machine.produce(100)
machine.produce(150)
machine.total_production()              
              
                       
```
